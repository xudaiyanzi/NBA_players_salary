```python
import os
import glob
import pandas as pd
os.chdir("/Users/Guobao/Dropbox/Bootcamp/Projects/3/NBA_players_salary/regression/dataset for Tableau/method2")
```

```python
# use glob to match the pattern 'csv'
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
method2_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
```
