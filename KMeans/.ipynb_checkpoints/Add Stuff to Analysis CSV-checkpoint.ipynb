```python
import pandas as pd
```

```python
DF = pd.read_csv('clusters_and_salary.csv', index_col = 0)
statsDF = pd.read_csv('Seasons_Stats.csv', index_col = 0)
```

```python
DF.head()
```

```python
statsDF.head()
```

```python
statsDF = statsDF[['Year', 'Player', 'Pos', 'Tm']]
```

```python
statsDF = statsDF.dropna()
```

```python
statsDF['Year'] = statsDF['Year'].astype(int)
```

```python
statsDF.head()
```

```python
def yearFilter(row):
    if not row['Year']:
        return
    if row['Year'] < 2010:
        return 
    elif row['Year'] > 2016:
        return
    else:
        return int(row['Year'])
statsDF['Year'] = statsDF.apply(yearFilter, axis = 1)
statsDF = statsDF.dropna()
statsDF.head()
```

```python
statsDF['Pos'].value_counts()
```

```python
statsDF = statsDF.rename({'Player' : 'Name'}, axis = 'columns')
```

```python
DF = DF.merge(statsDF, on = ['Year', 'Name'], how = 'left')
```

```python
DF.head(50)
```

```python
testDF = DF.groupby(['Year', 'Name'])
testDF.count()
```

```python
###### test without team
smallStatsDF = statsDF.drop('Tm', axis = 'columns')
DF = DF.merge(smallStatsDF, on = ['Year', 'Name'], how = 'left')
DF.head()
```

```python
DF.to_csv('cluster_pos_tm.csv')
```
