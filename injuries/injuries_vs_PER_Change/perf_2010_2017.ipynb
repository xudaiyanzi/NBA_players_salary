```python
# import Pandas
import pandas as pd
```

```python
df_1 = pd.read_csv("Seasons_Stats.csv")
df_1.head()
```

```python
# Clean up the player stats data to get ready for merging
# keep useful columns
df_2 = df_1[["Year", "Player","G","Pos","Age","PER", "Tm"]]
```

```python
# Get rid of NaN values
df_2 = df_2.dropna(subset =["Year"])
df_2.Year = df_2.Year.astype(int)
```

```python
# rname column
df_2 = df_2.rename(columns={"Player": "Name","G":"GP"})
df_2.GP = df_2.GP.astype(int)
```

```python
# Delete the * in the name column to keep consistant
df_2.Name = df_2.Name.astype(str)
```

```python
# Delete the * in the name column to keep consistant
df_2["Name"] = df_2["Name"].apply(lambda x: x.replace("*", ""))
```

```python
# it is import to say the year after 1980 gave more complete data
## here we are using data after 2009 due to the data avalability of injuries

df_3 = df_2.loc[df_2["Year"] >2009]
```

```python
df_4 = df_3.loc[df_3["Year"] <2019]
```

```python
df_4.head()
```

```python
# some team changed the name
def rename_team(x,Tm):
    if x[Tm] == "NOH":
        x[Tm] = "NOP"
    elif x[Tm] == "CHO":
        x[Tm] = "CHA"
    elif x[Tm] == "NJN":
        x[Tm] = "BRK"
    return x
    
df_4 = df_4.apply(lambda x: rename_team(x,"Tm"), axis=1)
```

```python
df_4["Tm"].unique()
```

```python
# Get rid of NaN values
player_stats_df = df_4.dropna()
```

```python
## this code is just to check whether we over deleted the data
# df_3.loc[df_3["Name"] == "Yao Ming"]
```

```python
# change the team name
team_name = pd.read_fwf('NBA_team_name.txt', header=None)
#team_name
```

```python
team_name["city_1"], team_name["team_1"] = team_name[0].str.split(' ', 1).str
team_name["city_2"] = team_name["city_1"].astype(str).str[:3]
team_name["city_2"] = team_name["city_2"].str.upper()

## change the LA and NEW city name
team_name["city_2"][2] = "BRK"
team_name["city_2"][3] = "NOP"
team_name["city_2"][9] = "GSW"
team_name["city_2"][12] = "LAC"
team_name["city_2"][13] = "LAL"
team_name["city_2"][19] = "NYK"
team_name["city_2"][20] = "OKC"
team_name["city_2"][26] = "SAS"

## change the team city name
team_name["team_1"][3] = "Pelicans"
team_name["team_1"][9] = "Warriors"
team_name["team_1"][19] = "Knicks"
team_name["team_1"][20] = "Thunder"
team_name["team_1"][24] = "Blazers"
team_name["team_1"][26] = "Spurs"

team_final_1 = team_name[["team_1","city_2"]]
team_final = team_final_1.rename(columns={'team_1': 'Team_name', 'city_2': 'Tm'})
```
