```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
injury_per = pd.read_csv('injury_per_change.csv')
injury_per_1 = injury_per.drop(["Injury Total"], axis=1)
injury_per_1.head()
```

```python
# X = injury_per[['GP', 'Age', 'PER':-1]]
X = injury_per_1.iloc[:,2:-1]
y = injury_per_1["PER_change"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30)
```

# Model begin!!!!

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
len(y_test)
```

```python
# from keras.utils import to_categorical

# One-hot encoding
# y_train_categorical = to_categorical(y_train)
# y_test_categorical = to_categorical(y_test)
```

```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=100, activation='tanh', input_dim=40))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=100, activation='relu'))
# model.add(Dense(units=1, activation='softmax'))
model.add(Dense(units=1, activation='relu'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='mse',
              metrics=['accuracy', 'mse'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train,
    epochs=200,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy, model_mse = model.evaluate(
    X_test_scaled, y_test, verbose=40)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}, MSE: {model_mse}")
```
