```python
# import Pandas
import pandas as pd
```

```python
# read performance data from 2010_2017
perf_df = pd.read_csv("perf_2010_2017.csv")

perf = perf_df[["Year","Name","GP","Pos","Age","PER"]]
perf.count()
```

```python
# read pysical data from 2010_2017
inj_df = pd.read_csv("injury_clean_details.csv")
inj_df.count()
```

```python
# merge performance stat and physical stats as perf_phy_df
#### we lost data due to inner join
perf_inj_df = pd.merge(perf, inj_df, on=["Year","Name"],how="left")
perf_inj_df.count()
```

```python
# here we do left join and fillna with 0 because not all the players are injured 
perf_inj_df = perf_inj_df.fillna(0)
perf_inj_df.count()
```

```python
df = perf_inj_df.sort_values(['Name','Year'])
```

```python
# may be do average is easier
df_ave = df.groupby(['Year', 'Name']).mean()
df_ave = df_ave.reset_index(level=[0,1])
```

```python
df_ave_sort = df_ave.sort_values(['Name','Year'])
```

```python
df_ave_sort_reset = df_ave_sort.reset_index(drop = True)
df_ave_sort_reset.head(10)
```

```python
## save clean_merge data
df_ave_sort_reset.to_csv("perf_inj_details_ave_sort.csv", index = False)
```

```python
df_next_per_1 = df_ave_sort_reset
# this change is from previous year, default is 0
df_next_per_1["PER_change"] = 0
df_next_per_1.head()
```

```python
for i in range(len(df_next_per_1)-1):
    
    next_year = df_next_per_1.loc[i+1,"Year"]
    next_name = df_next_per_1.loc[i+1,"Name"]
    
    if df_next_per_1.loc[i, "Year"] == (next_year-1) and df_next_per_1.loc[i, "Name"] == next_name:
        # how this year injury change PER in next year
        df_next_per_1.loc[i, "PER_change"] = df_next_per_1.loc[i+1, "PER"] - df_next_per_1.loc[i, "PER"]
```
