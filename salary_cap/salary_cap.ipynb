```python
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
#Salary cap data for the year 1984-2017. The https://www.basketball-reference.com/contracts/salary-cap-history.html
csvpth = "Salary_cap.csv"
salary_cap_df = pd.read_csv(csvpth)
salary_cap_df
```

```python
salary_cap_df['Salary Cap'] = salary_cap_df['Salary Cap'].apply(lambda x: x.replace("$", ""))
salary_cap_df['Salary Cap'] = salary_cap_df['Salary Cap'].apply(lambda x: x.replace(",", ""))
salary_cap_df
```

```python
salary_cap_clean_df = salary_cap_df[['Year','Salary Cap']][15:]
salary_cap_clean_df
```

```python
salary_cap_clean_df['Year'] = salary_cap_clean_df['Year'].apply(lambda x: str('20'+x[-2:]))
salary_cap_clean_df.rename(index=str, columns={"Salary Cap": "SalaryCap"})
```
