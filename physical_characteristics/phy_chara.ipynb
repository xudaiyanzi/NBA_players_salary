```python
# import Pandas
import pandas as pd

# import os
import os

# import numpy
import numpy as np

# import pyplot
import matplotlib.pyplot as plt
```

### import csv file 

```python
physical_chara = pd.read_csv(os.path.join("all_seasons.csv"))
physical_chara.head()
```

### Restructuring the dataframe and change the season column format

```python
# change the order of columns and remove the same columns in stats data
phy_df = physical_chara[["season", "player_name","player_height","player_weight",'gp','net_rating']]
# rename columns
phy_df = phy_df.rename(columns={"player_name": "Name","age":"Age","season":"Season"})

# extract first 4 characters from Season column and set it as Year column
phy_df['Year'] = [x[:4] for x in phy_df['Season']]
# set the Year value as interger
phy_df.Year = phy_df.Year.astype(int)

phy_df.dtypes
```

```python
df = phy_df.groupby(['Year','Name']).mean()
df.head(50)
```
